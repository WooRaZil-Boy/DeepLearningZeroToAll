# Hypothesis not using matrix
<img src="Images/lab4_1.png">

```python
import tensorflow as tf

x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]

# placeholders for a tensor that will be always fed.
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]), name='weight1')
w2 = tf.Variable(tf.random_normal([1]), name='weight2')
w3 = tf.Variable(tf.random_normal([1]), name='weight3')
b = tf.Variable(tf.random_normal([1]), name='bias')

#tf.placeholder()는 입력 데이터를 만들 때 주로 사용한다. (실제 훈련 예제를 제공하는 변수) - 초기값을 지정할 필요 없다. (모델 입력시 변경되지 않을 데이터)
#tf.Variable()은 데이터의 상태를 저장할 때 주로 사용한다. (가중치나 편향 등의 학습 가능한 변수) - 초기값을 지정해야 한다. (학습 되는 데이터)
#http://stackoverflow.com/questions/36693740/whats-the-difference-between-tf-placeholder-and-tf-variable

hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b

# cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
# Minimize. Need a very small learning rate for this data set
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5) #0.00001
train = optimizer.minimize(cost)

# Launch the graph in a session.
with tf.Session() as session:
    # Initializes global variables in the graph.
    session.run(tf.global_variables_initializer())
    
    for step in range(2001):
        cost_val, hy_val, _ = session.run([cost, hypothesis, train], feed_dict={x1: x1_data, x2: x2_data, x3: x3_data, Y: y_data})
        
        if step % 10 == 0:
            print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
            
#결과값이 출력 되기는 하지만, 일일이 값을 입력하므로 비효율적이다.
#matrix를 쓰면 더 간단하게 정리할 수 있다.
```

# Hypothesis using matrix
<img src="Images/lab4_2.png">

```python
x_data = [[73., 80., 75.], [93., 88., 93.],
         [89., 91., 90.], [96., 98., 100.], [73., 66., 70.]] #행렬로 생성. 인풋 데이터
y_data = [[152.], [185.], [180.], [196.], [142.]] #행렬로 생성. 정답 레이블

# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, 3]) #shape=[None, 3] (개수 제한 없음(None), 각 요소는 3개의 값을 가진다)
Y = tf.placeholder(tf.float32, shape=[None, 1]) #shape=[None, 1] (개수 제한 없음(None), 각 요소는 1개의 값을 가진다)

W = tf.Variable(tf.random_normal([3, 1]), name='weight') #tf.random_normal 출력 텐서의 형태 (3, 1)
b = tf.Variable(tf.random_normal([1]), name='bias') #tf.random_normal 출력 텐서의 형태 1개의 랜덤 변수

# Hypothesis
hypothesis = tf.matmul(X, W) + b #행렬의 내적 + b
#tf.add(tf.matmul(X, W), b)와 같다

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5) #0.00001
train = optimizer.minimize(cost) #최적화

# Launch the graph in a session.
with tf.Session() as session:
    # Initializes global variables in the graph.
    session.run(tf.global_variables_initializer())
    
    for step in range(2001):
        cost_val, hy_val, _ = session.run([cost, hypothesis, train], feed_dict={X: x_data, Y: y_data})
        #실제 실행은 train, 학습의 진행 상황 알아보기 위한 변수 cost, hypothesis
        
        if step % 10 == 0:
            print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
```

```python
# Lab 4 Multi-variable linear regression
import numpy as np

xy = np.loadtxt('data-01-test-score.csv', delimiter=',', dtype=np.float32)
#numpy로 파일 불러오기 (파일 위치 + 이름, 분할단위, 데이터 타입 = float32)
#같은 데이터 타입인 경우에만 사용 가능하다.
x_data = xy[:, 0:-1] #전체 열을 가져온다, 각 열의 0 ~ index-1 행까지 가져온다.
y_data = xy[:, [-1]] #전체 열을 가져온다, 각 열의 마지막 행만 가져온다.

# Make sure the shape and data are OK
print(x_data.shape, x_data, len(x_data))
print(y_data.shape, y_data)
```

```python
# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, 3]) #데이터와 shape를 일치시킨다.
Y = tf.placeholder(tf.float32, shape=[None, 1]) #정답레이블과 shape를 일치시킨다.
#placeholder는 입력값, 정답 레이블 등 바뀌지 않는 상수. session.run()시에 값 입력

W = tf.Variable(tf.random_normal([3, 1]), name='weight') #들어오는 값 3, 나가는 값 1. 3x1 행렬
b = tf.Variable(tf.random_normal([1]), name='bias') #나가는 값 1
#Variable은 가중치, 편향 등 바뀌는 변수

# Hypothesis
hypothesis = tf.matmul(X, W) + b #행렬 내적곱

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y)) #오차 제곱의 평균

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5) #학습률 0.00005. 손실함수 GeadientDescent
train = optimizer.minimize(cost) #코스트 최적화

# Launch the graph in a session.
with tf.Session() as session:
    session.run(tf.global_variables_initializer()) #초기화
    
    for step in range(2001):
        cost_val, hy_val, _ = session.run([cost, hypothesis, train], feed_dict={X: x_data, Y: y_data})
        
        if step % 10 == 0: #10번 마다 출력
            print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
            
    # Ask my score
    #학습된 모델로 예측하기
    print("Your score will be ", session.run(hypothesis, feed_dict={X: [[100, 70, 101]]}))
    print("Other scores will be ", session.run(hypothesis, feed_dict={X: [[60, 70, 110], [90, 100, 80]]}))
    #학습이 아닌 단순 예측이므로 hypothesis만으로 run 하면 된다.
```

# Queue Runners

<img src="Images/lab4_4.png">
불러올 데이터가 매우 큰 경우 numpy에서 out of memory 오류가 날 수 있다.
tf의 Queue를 이용해서 불러올 수 있다.

<img src="Images/lab4_3.gif">

```python
filename_queue = tf.train.string_input_producer(['data-01-test-score.csv'], shuffle=False, name='filename_queue')
#1. 불러올 파일들의 리스트를 정리한다.

reader = tf.TextLineReader()
key, value = reader.read(filename_queue)
#2. TextLineReader를 정의해 파일을 로드한다.

# Default values, in case of empty columns. Also specifies the type of the
# decoded result.
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)
#3. Data를 가져온다.

# collect batches of csv in
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10) 
#[xy[0:-1], xy[-1:]] : x데이터, y데이터, 한번에 10개씩(batch_size) 가져온다.

# placeholders for a tensor that will be always fed.
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# Hypothesis
hypothesis = tf.matmul(X, W) + b

# Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

# Launch the graph in a session.
with tf.Session() as session:
    # Initializes global variables in the graph.
    session.run(tf.global_variables_initializer())
    
    # Start populating the filename queue.
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(sess=session, coord=coord) #큐를 시작
    
    for step in range(2001):
        x_batch, y_batch = session.run([train_x_batch, train_y_batch]) #배치또한 텐서 이므로 run이 필요하다. 
        cost_val, hy_val, _ = session.run([cost, hypothesis, train], feed_dict={X: x_batch, Y: y_batch})
        #가져온 배치를 feed_dict에 사용한다.
        
        if step % 10 == 0:
            print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
            
    
    coord.request_stop() #큐 종료
    coord.join(threads)
    
    # Ask my score
    print("Your score will be ", session.run(hypothesis, feed_dict={X: [[100, 70, 101]]}))
    print("Other scores will be ", session.run(hypothesis, feed_dict={X: [[60, 70, 110], [90, 100, 80]]}))
```
